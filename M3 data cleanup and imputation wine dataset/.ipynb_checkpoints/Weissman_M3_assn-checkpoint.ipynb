<br> <br>

**In reviewing the missing data:**
- We've found that 'STARS' cannot be used to make predictions due to the sheer amount of data missing for it
- We've found that we should be careful if/when imputing for 'Sulphates', as we don't have data for close to 8% of observations

We check to see if any data rows are missing values in a lot of dimensions at once, and decide to remove these outright before proceeding. The threshold we decide on is 25% (4 or more columns missing).

```python
droplist = []
for i in range(raw_data.shape[0]):
    missing = raw_data.iloc[i].isna().sum()
    if missing >= 0.25*raw_data.shape[1]:
        droplist.append(i)

len(droplist)
```

```python
raw_data.drop(droplist, axis = 0, inplace = True)
raw_data.reset_index(inplace = True)
raw_data.drop('index', axis = 1, inplace = True)

raw_data.shape
```

Less than 0.2% percent of obeservations (25/12700) are missing 4 or more attributes, so their removal is unlikely to make a significant impact on our EDA.


Helper functions for EDA:

```python
def eda_plot(df, col):
    '''Generate histogram and box plot for a given numeric attribute.'''
    plt.figure(figsize = (10,4))
    
    plt.subplot(1,2,1)
    plt.title('Histogram of ' + col)
    sns.histplot(x = df[col], kde = True)
    plt.subplot(1,2,2)
    plt.title('Box Plot of ' + col)
    sns.boxenplot(x = df[col])
    
    plt.tight_layout()
    plt.show()

def eda_cat_plot(df, col):
    '''Generate count plot for a given categorical attribute.'''
    plt.figure(figsize = (6,4))
    
    plt.title('Count of ' + col)
    sns.countplot(x = col, data = df)
    
    plt.tight_layout()
    plt.show()

def eda_modify(df_old, df_new, col):
    '''Compare before-and-after histograms and box plots for a modified numeric attribute.'''
    plt.figure(figsize = (6,4))
    
    plt.subplot(2,2,1)
    plt.title('Old: ' + col)
    sns.histplot(x = df_old[col], kde = True)
    plt.subplot(2,2,2)
    sns.boxenplot(x = df_old[col])
    
    plt.subplot(2,2,3)
    plt.title('New: ' + col)
    sns.histplot(x = df_new[col], kde = True)
    plt.subplot(2,2,4)
    sns.boxenplot(x = df_new[col])
    
    plt.tight_layout()
    plt.show()
```

# Data Preparation & Modified Data Review

## Step 1: Numeric Missing Value Imputation

The first action we take to prepare our data is to impute missing numerical values, since we cannot use machine learning on a data set that contains missing values. We use a KNN imputaton method from the scikit-learn library and store the transformed dataframe separately. We select k = 10 instead of the default of k = 5 so as to incorporate more neighborhood data at the cost of slightly more computational resources. Our data set is small so computing power is not a limitation.

```python
from sklearn.impute import KNNImputer
```

```python
numerics = ['TARGET', 'FixedAcidity', 'VolatileAcidity', 'CitricAcid',
            'ResidualSugar', 'Chlorides', 'FreeSulfurDioxide',
            'TotalSulfurDioxide', 'Density', 'pH', 'Sulphates', 'Alcohol', 'AcidIndex']

knn = KNNImputer(n_neighbors = 10)

filled_data = knn.fit_transform(raw_data[numerics])

filled_data.shape
```

```python
filled_data = pd.DataFrame(filled_data, columns = numerics)

filled_data
```

## Step 2: Numeric Power Transform

The second step we conduct is a power transform on the numeric component on our data set. Since none of our numeric variables are normally distributed and a lot of them are highly skewed, we need to transform them to avoid creating biased models.

Once again, we use a scikit-learn preprocessing library to perform this step.

```python
from sklearn.preprocessing import power_transform
```

```python
filled_data.describe()
```

We note that the data contains zero-valued entries in almost every attribute, and is thus not strictly positive for a Box-Cox transform. We use a similar Yeo-Johnson transform instead, which is able to handle zero-valued data.

```python
transformed_data = pd.DataFrame(power_transform(filled_data, method = 'yeo-johnson'), columns = numerics)

transformed_data
```

## Step 3: Final Reintegration

Finally, we reintegrate our transformed numeric data with our categorical data. The categorical fields are converted into dummy variables for use in machine learning. STARS is still included in this data set but we do not recommend its usage as a predictor. Instead its missing data has been sequestered into its own indicator column (STARS_Unknown) as a result of the dummy variable transformation.

```python
final_data = transformed_data.join(pd.get_dummies(raw_data['STARS'], prefix='STARS'))
final_data = final_data.join(pd.get_dummies(raw_data['LabelAppeal'], prefix='LblAppeal'))

final_data
```

# Conclusion

In conclusion, we have put forth an effort towards converting a messy data set that is not suitable for machine learning applications into one that is relatively useful.

The procedure for handling missing data in numeric variables via imputation was successful, preserving the shape of the distributions and maintaining the volume of data available to us. Nothing useful could be done for the categorical variable STARS however because of the very large proportion of data missing. The categorical variables were converted to one-hot dummy indicators for further use in machine learning.

The attempt to transform the shape of our numeric attributes was also noteworthy. Our use of a Yeo-Johnson transform was able to nullify heavy skew in many of our numeric distributions by converting them into multimodal distributions, as well as rescaling them to a common scale.
