```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def uneven_die_sampler(n):
    # we define a biased die
    #first we deifne a probability density function
    pdf = [(1, 2/6), (2, 0), (3, 1/6), (4, 1/6), (5, 0), (6, 2/6)]
    #then we define a cumilitive distribution funtion
    cdf = [(i, sum(p for j,p in pdf if j < i)) for i,_ in pdf]
    #based on the pfd and cdf, we pick a random number
    R = max(i for r in [random.random()] for i,c in cdf if c <= r)
    sample = []
    for _ in range(n):
        datum = max(i for r in [random.random()] for i,c in cdf if c <= r)
        sample.append(datum)
#     print(sample)
    return np.array(sample)
# np.mean(uneven_die_sampler(10))
```

```python
#beware! this demo is by no means efficent and some of the code can definitely be optimized
# it is only meant to demonstrate the CLT
sample_mean_dict = {}


# we're run the uneven die sampler 
# we use a predifned sample size 
# and run the uneven_die_sampler a bunch of times
for _ in range(5000):
    sample = uneven_die_sampler(200)
    mean = np.mean(sample)
    #here we add/incr the # occurences for the current mean
    if mean in sample_mean_dict:
        sample_mean_dict[mean] += 1
    else:
        sample_mean_dict[mean] = 1
# x,y lists which we will used to plot distribution of the above experiment
x,y = [], []
#sort the dict based on mean frequncy (value)
sample_mean_dict = sorted(sample_mean_dict.items(), key = lambda kv:(kv[0], kv[1]))
for key,val in sample_mean_dict:
    x.append(key)
    y.append(val)
```

```python
plt.plot(x,y)
plt.title('Mean Frequency')
plt.show()
```

```python
#beware! this demo is by no means efficent and some of the code can definitely be optimized
# it is only meant to demonstrate the CLT
sample_sum_dict = {}


# we're run the uneven die sampler 
# we use a predifned sample size 
# and run the uneven_die_sampler a bunch of times
for _ in range(50000):
    sample = uneven_die_sampler(89)
    sample_sum = np.sum(sample)
    #here we add/incr the # occurences for the current sum
    if sample_sum in sample_sum_dict:
        sample_sum_dict[sample_sum] += 1
    else:
        sample_sum_dict[sample_sum] = 1
# x,y lists which we will used to plot distribution of the above experiment
x,y = [], []
#sort the dict based on sum frequncy (value)
sample_sum_dict = sorted(sample_sum_dict.items(), key = lambda kv:(kv[0], kv[1]))
for key,val in sample_sum_dict:
    x.append(key)
    y.append(val)
```

```python
plt.plot(x,y)
plt.title('Sum Frequency')
plt.show()
```
