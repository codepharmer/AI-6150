### Introduction

As of recently, Americans spend close to 9% of their disposable income on food$^1$
. Additionally, according to data from the Bureau of Labor Statistics, the US has been facing extreme inflation on food prices in 2022$^2$
. With all the inflation, food pricing has been on our minds recently. With this in mind, we managed to procure food pricing data from kosher supermarkets in NYC which we will focus on in this research project.

### Research Questions

In completing this project, we hope to gain a better understanding of food pricing and share that with the reader. We want to find the driving factors of item pricing and understand our categorical variables and which items are mostly likely to be on sale. We also aim to build a model which can predict item-price using four or five explanatory variables.


This type of data may be helpful in informing consumers of where they are likely to find the best priced produce/groceries based on different metrics. We can also use the data to help store owners better understand their own pricing models and how their models compare with the pricing models of their competitors.

### Data To Be Used

Supermarket data being used was scraped from websites of [three](https://www.moishasonline.com/) [different](https://www.thepompeopleonline.com/) [supermarkets](https://www.glattmart.com/) (linked here and then again at the bottom).

Each observation contains:

- store - categorical nominal

- weight – composed of numerical and potential categorical data
- sale-price – numerical continuous
- category - categorical nominal
- main-category - categorical nominal
- date – date (does not necessarily fit the typical typology)
- sub-category - categorical nominal

### Approach

We explained where the data is coming from earlier.

While some data cleansing will likely be necessary, if we end up with too many categories in some of our fields, we may also need to aggregate some of our categorical data.

In our data analysis we will want to see how prices varies by brand category, store etc. We will want to know which stores offer the most competitive prices for each category and we may want to consider pricing variance by category.

As we are working with a dataset that has few explanatory variables, feature reduction may not be necessary, but that is to be determined. We may use some feature reduction methods just to double check.

As far as ML algorithms, with this amount of categorical data, we will likely want to consider random forest; KNN and logistic regression. 
We may also want to use neural networks to determine categories for a given item.

We then can compare the accuracy of our KNN, logistic regression and ANN models, to deterimine which approach would be most applicable with this dataset.

We can then build an SVM and a decision tree classifier to use in an ensemble model to predict categorical values for our given input.
